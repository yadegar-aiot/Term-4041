<div dir="rtl" style="text-align:right;"><h2 style="margin-top:0;">1) حلقه‌ها (Loops)</h2>حلقه یعنی «تکرار یک کار»؛ یا تا وقتی یک شرط برقرار است ادامه می‌دهیم، یا روی یک مجموعه داده حرکت می‌کنیم.<br><br><h3 style="margin:0;">1-1) حلقه <code class="ltr">for</code></h3>وقتی می‌خواهیم روی یک دنباله مثل <code class="ltr">range</code>، لیست، رشته یا دیکشنری حرکت کنیم از <code class="ltr">for</code> استفاده می‌کنیم.<br></div>

```python
# for + range (شمارنده)
for i in range(1, 6):   # 1 تا 5
    print(i)

# range با گام (step)
for i in range(2, 11, 2):   # 2,4,6,8,10
    print(i)

# for روی لیست
for n in [10, 20, 30]:
    print(n)

# for روی رشته
for ch in "Python":
    print(ch)
```

<div dir="rtl" style="text-align:right;"><h3 style="margin-top:0;">1-2) حلقه <code class="ltr">while</code></h3>وقتی تعداد تکرار دقیقاً معلوم نیست و «تا وقتی شرط درست است» ادامه می‌دهیم، <code class="ltr">while</code> انتخاب خوبی است.<br><br><b>ترفند مهم:</b> اگر بخواهیم «حتماً حداقل یک‌بار اجرا شود»، معمولاً از <code class="ltr">while True</code> استفاده می‌کنیم و با <code class="ltr">break</code> از حلقه خارج می‌شویم (شبیه ایدهٔ <span class="ltr">do...while</span> در بعضی زبان‌ها).<br></div>

```python
# while True (شبیه do...while): جمع اعداد تا وقتی کاربر 0 نزده
total = 0
while True:
    n = int(input("عدد بده (0 برای پایان): "))
    if n == 0:
        break
    total += n

print("مجموع:", total)
```

<div dir="rtl" style="text-align:right;"><h3 style="margin-top:0;">1-3) <code class="ltr">break</code> و <code class="ltr">continue</code></h3><ul><li><code class="ltr">break</code> یعنی «همین الان حلقه را تمام کن».</li><li><code class="ltr">continue</code> یعنی «این دور را رد کن و برو دور بعد».</li></ul><h3 style="margin-top:12px;">1-4) ابزارهای خیلی کاربردی: <code class="ltr">enumerate</code> و <code class="ltr">zip</code></h3><b><code class="ltr">enumerate</code> چه می‌کند؟</b> وقتی می‌خواهیم همراه با هر آیتم، شمارهٔ آن را هم داشته باشیم؛ خروجی‌اش «جفتِ (index, value)» است.<br>می‌توانیم شماره‌گذاری را با <code class="ltr">start=1</code> از ۱ شروع کنیم.<br><b><code class="ltr">zip</code> چه می‌کند؟</b> چند دنباله را «کنار هم» پیمایش می‌کند و در هر دور یک «تاپلِ ترکیبی» می‌دهد؛ یعنی مثل زیپ دو تا لیست را جفت‌جفت جلو می‌برد.<br><blockquote><b>نکته:</b> <code class="ltr">zip</code> تا طولِ کوتاه‌ترین دنباله جلو می‌رود؛ اگر طول‌ها برابر نیست، اضافه‌ها نادیده گرفته می‌شوند.</blockquote></div>

```python
names = ["Ali", "Sara", "Reza"]

# enumerate: شماره + مقدار
for idx, name in enumerate(names, start=1):
    print(idx, name)

# zip: حرکت همزمان روی دو لیست
scores = [18, 20, 15]
for name, score in zip(names, scores):
    print(name, score)
```

<div dir="rtl" style="text-align:right;"><b>چند مثال تکمیلی:</b><br>۱) پیمایش روی دیکشنری معمولاً با <code class="ltr">items()</code> انجام می‌شود تا هم کلید را داشته باشیم هم مقدار را.<br>۲) در حلقه‌ها مراقب «تورفتگی (indentation)» باشید؛ یک فاصلهٔ اضافه یا کم می‌تواند <code class="ltr">IndentationError</code> بدهد.<br><br><h3 style="margin-top:0;">خطاهای رایج در حلقه‌ها</h3><ul><li><b>حلقهٔ بی‌نهایت:</b> در <code class="ltr">while</code> شرط هیچ‌وقت false نمی‌شود یا مقدار آپدیت نمی‌شود.</li><li><b>اشتباه در <code class="ltr">range</code>:</b> <code class="ltr">range(a,b)</code> شامل <span class="ltr">b</span> نیست (خطای یک‌دونه جلو/عقب).</li><li><b>دستکاری لیست هنگام پیمایش:</b> حذف/افزودن آیتم داخل همان حلقه می‌تواند نتیجهٔ عجیب بدهد؛ بهتر است روی یک کپی پیمایش کنید.</li><li><b>فراموش کردن unpack:</b> در <code class="ltr">for idx, val in enumerate(lst)</code> باید دو متغیر داشته باشید؛ وگرنه <code class="ltr">ValueError</code> می‌گیرید.</li></ul></div>

```python
# for روی دیکشنری: کلید + مقدار
d = {"a": 1, "b": 2, "c": 3}
for k, v in d.items():
    print(k, v)

print("----")

# for ... else: جستجو
nums = [2, 4, 6, 8]
target = 5

for n in nums:
    if n == target:
        print("پیدا شد:", n)
        break
else:
    print("پیدا نشد!")

print("----")

# حلقهٔ تو در تو (نمونه کوچک)
for i in range(1, 4):
    for j in range(1, 4):
        print(i * j, end=" ")
    print()
```

<div dir="rtl" style="text-align:right;"><h2 style="margin-top:0;">2) لیست و تاپل</h2><h3 style="margin:0;">2-1) لیست (<code class="ltr">list</code>)</h3>لیست یک مجموعهٔ «قابل تغییر» است؛ می‌توانیم آیتم اضافه/حذف کنیم یا مقدار یک خانه را عوض کنیم.<br></div>

```python
nums = [10, 20, 30]
nums.append(40)
nums[0] = 99
print(nums)
```

<div dir="rtl" style="text-align:right;"><h3 style="margin-top:0;">2-2) تاپل (<code class="ltr">tuple</code>)</h3>تاپل «غیرقابل تغییر» است؛ برای داده‌های ثابت (مثل مختصات، خروجی چندتایی تابع، یا رکوردهای ثابت) انتخاب خوبی است.<br><blockquote><b>نکته:</b> تاپل تک‌عضوی باید کاما داشته باشد: <code class="ltr">(5,)</code> وگرنه فقط یک عدد داخل پرانتز می‌شود.</blockquote></div>

```python
point = (3, 5)
# point[0] = 10  # خطا: قابل تغییر نیست
print(point)
```

<div dir="rtl" style="text-align:right;"><h3 style="margin-top:0;">2-3) تفاوت‌های مهم لیست و تاپل</h3><ul><li><b>لیست:</b> قابل تغییر، مناسب داده‌های پویا.</li><li><b>تاپل:</b> غیرقابل تغییر، مناسب داده‌های ثابت و خروجی چندتایی.</li></ul><h3 style="margin-top:12px;">2-4) کوتاه‌نویسی پایتونی: List Comprehension</h3>یک راه تمیز برای ساخت لیست‌ها در یک خط (با امکان فیلتر کردن).<br><br><h3 style="margin-top:0;">خطاهای رایج در لیست/تاپل</h3><ul><li><b><code class="ltr">IndexError</code>:</b> اندیس خارج از محدوده (مثلاً <code class="ltr">lst[10]</code> وقتی لیست کوتاه‌تر است).</li><li><b>اشتباه در <code class="ltr">append</code>:</b> <code class="ltr">append</code> چیزی برنمی‌گرداند؛ اگر بنویسید <code class="ltr">lst = lst.append(x)</code> لیست‌تان <code class="ltr">None</code> می‌شود.</li><li><b>کپی vs ارجاع:</b> <code class="ltr">b = a</code> کپی نیست؛ هر دو به یک لیست اشاره می‌کنند. برای کپی از <code class="ltr">a.copy()</code> یا <code class="ltr">a[:]</code> استفاده کنید.</li><li><b>اشتباه در تاپل تک‌عضوی:</b> <code class="ltr">(5)</code> تاپل نیست؛ باید <code class="ltr">(5,)</code> باشد.</li></ul></div>

```python
# ساخت لیست مربع‌ها
squares = [x*x for x in range(1, 6)]
print(squares)

# فیلتر: فقط زوج‌ها
evens = [x for x in range(1, 11) if x % 2 == 0]
print(evens)
```

<div dir="rtl" style="text-align:right;"><h2 style="margin-top:0;">3) دیکشنری (<code class="ltr">dict</code>)</h2>دیکشنری یعنی نگاشت «کلید → مقدار». مثل دفترچه تلفن: نام → شماره.<br><blockquote><b>نکته:</b> کلیدها باید «تغییرناپذیر» باشند (مثل رشته/عدد/تاپل). لیست نمی‌تواند کلید باشد.</blockquote></div>

```python
person = {"name": "Ali", "age": 21}

print(person["name"])        # دسترسی مستقیم
print(person.get("city"))    # اگر نبود None می‌دهد (خطا نمی‌دهد)

person["city"] = "Tehran"    # افزودن
person["age"] = 22           # تغییر
print(person)
```

<div dir="rtl" style="text-align:right;"><h3 style="margin-top:0;">پیمایش روی دیکشنری</h3><ul><li>پیمایش مستقیم روی دیکشنری یعنی پیمایش روی <b>کلیدها</b>.</li><li>برای گرفتن همزمان کلید و مقدار از <code class="ltr">items()</code> استفاده می‌کنیم.</li></ul></div>

```python
for key in person:
    print(key, person[key])

for key, value in person.items():
    print(key, "=>", value)
```

<div dir="rtl" style="text-align:right;"><h3 style="margin-top:0;">Dict Comprehension (پیشنهادی)</h3>مثل لیست‌کامپرهنشن، ولی خروجی دیکشنری می‌سازد (مثلاً عدد → مربع).<br><br><h3 style="margin-top:0;">خطاهای رایج در دیکشنری</h3><ul><li><b><code class="ltr">KeyError</code>:</b> دسترسی به کلیدی که وجود ندارد (به‌جای آن از <code class="ltr">get</code> استفاده کنید یا قبلش <code class="ltr">in</code> چک کنید).</li><li><b>فکر کردن <code class="ltr">in</code> روی مقادیر است:</b> <code class="ltr">"Ali" in d</code> کلیدها را چک می‌کند، نه مقدارها.</li><li><b>بازنویسی ناخواسته:</b> اگر دوباره همان کلید را ست کنید، مقدار قبلی overwrite می‌شود.</li><li><b>کلید نامعتبر:</b> استفاده از لیست به‌عنوان کلید خطا می‌دهد: <code class="ltr">TypeError: unhashable type: list</code>.</li></ul></div>

```python
# عدد -> مربع
sq = {x: x*x for x in range(1, 6)}
print(sq)
```

<div dir="rtl" style="text-align:right;"><h2 style="margin-top:0;">4) مدیریت خطا با <code class="ltr">try</code></h2>برنامه‌های واقعی همیشه ورودی عجیب می‌گیرند یا خطاهای غیرمنتظره دارند؛ ما باید «مدیریت‌شان» کنیم تا برنامه ناگهانی کرش نکند.<br></div>

```python
try:
    x = int(input("یک عدد وارد کن: "))
    print("100 تقسیم بر x =", 100 / x)
except ValueError:
    print("خطا: لطفاً عدد معتبر وارد کن.")
except ZeroDivisionError:
    print("خطا: تقسیم بر صفر مجاز نیست.")
```

<div dir="rtl" style="text-align:right;"><h3 style="margin-top:0;"><code class="ltr">else</code> و <code class="ltr">finally</code></h3><ul><li><code class="ltr">else</code>: وقتی هیچ خطایی رخ نداد اجرا می‌شود.</li><li><code class="ltr">finally</code>: چه خطا بشود چه نشود اجرا می‌شود (پاکسازی/بستن منابع).</li></ul></div>

```python
try:
    n = int(input("عدد: "))
except ValueError:
    print("عدد نبود!")
else:
    print("عدد معتبر بود:", n)
finally:
    print("پایان این بخش (finally).")
```

<div dir="rtl" style="text-align:right;"><blockquote><b>نکته مهم:</b> از <code class="ltr">except:</code> خالی (بدون نوع خطا) فقط وقتی استفاده کن که واقعاً می‌دانی چه می‌کنی؛ چون همه‌چیز را قورت می‌دهد و دیباگ را سخت می‌کند.</blockquote><h3 style="margin-top:0;">خطاهای رایج در try/except</h3><ul><li><b>بلعیدن خطا:</b> خطا را می‌گیریم ولی هیچ پیام/لاگی نمی‌دهیم؛ بعداً نمی‌فهمیم چه شد.</li><li><b>ترتیب غلط exceptها:</b> اگر <code class="ltr">Exception</code> را اول بگذارید، بقیهٔ exceptها بی‌اثر می‌شوند.</li><li><b>گرفتن خطای اشتباه:</b> مثلاً برای تبدیل عدد باید <code class="ltr">ValueError</code> را بگیریم، نه <code class="ltr">TypeError</code>.</li><li><b>فقط try کردن همه‌چیز:</b> گاهی بهتر است قبل از try، ورودی را با شرط‌ها چک کنیم تا کنترل دست خودمان باشد.</li></ul></div>

<div dir="rtl" style="text-align:right;"><h2 style="margin-top:0;">5) کار با کتابخانه‌ها و <code class="ltr">import</code></h2>کتابخانه‌ها یعنی قابلیت‌های آماده. چند شکل رایج import:<br><ul><li><code class="ltr">import math</code></li><li><code class="ltr">from math import sqrt</code></li><li><code class="ltr">import numpy as np</code> (نام مستعار)</li></ul></div>

```python
import math
print(math.sqrt(16))
print(math.pi)

from random import randint
print(randint(1, 6))
```

<div dir="rtl" style="text-align:right;"><blockquote><b>نکته:</b> برای نصب کتابخانه‌های خارج از پایتون (مثل <span class="ltr">numpy</span>) معمولاً از <span class="ltr">pip</span> استفاده می‌شود؛ اما «نصب» با «<code class="ltr">import</code>» فرق دارد.</blockquote><h3 style="margin-top:0;">خطاهای رایج در import و کتابخانه‌ها</h3><ul><li><b><code class="ltr">ModuleNotFoundError</code>:</b> کتابخانه نصب نیست یا محیط (venv/conda) اشتباه است.</li><li><b>سایه انداختن روی نام کتابخانه:</b> اگر فایل‌تان را مثلاً <code class="ltr">random.py</code> نام‌گذاری کنید، ممکن است با ماژول <span class="ltr">random</span> تداخل کند.</li><li><b>استفاده از <code class="ltr">import *</code>:</b> خوانایی را کم می‌کند و تعارض اسم می‌سازد.</li><li><b>اشتباه در نام مستعار:</b> اگر <code class="ltr">import numpy as np</code> کردید، بعداً باید <code class="ltr">np</code> را صدا بزنید نه <code class="ltr">numpy</code>.</li></ul></div>

<div dir="rtl" style="text-align:right;"><h2 style="margin-top:0;">6) تابع (Function)</h2>تابع یعنی یک «قطعه کد قابل‌استفادهٔ مجدد» برای یک کار مشخص؛ کد را تمیز می‌کند، تکرار را کم می‌کند و تست را آسان‌تر می‌کند.<br></div>

```python
def add(a, b):
    return a + b

print(add(3, 4))
```

<div dir="rtl" style="text-align:right;"><h3 style="margin-top:0;">تفاوت <code class="ltr">print</code> و <code class="ltr">return</code></h3><ul><li><code class="ltr">print</code> فقط نمایش می‌دهد.</li><li><code class="ltr">return</code> خروجی را برمی‌گرداند تا بعداً در جای دیگر استفاده کنیم.</li></ul><h3 style="margin-top:12px;">پارامتر پیش‌فرض (Default Parameter)</h3>می‌توانیم برای ورودی تابع مقدار پیش‌فرض تعیین کنیم تا اگر کاربر چیزی نداد، از همان استفاده شود.<br></div>

```python
def greet(name="دوست من"):
    return f"سلام {name}!"

print(greet())
print(greet("پوریا"))
```

<div dir="rtl" style="text-align:right;"><h3 style="margin-top:0;">Docstring برای توضیح تابع</h3>اگر اولین خط داخل تابع یک رشتهٔ سه‌کوتیشن باشد، به عنوان «مستندات تابع» شناخته می‌شود و در <code class="ltr">help()</code> و <code class="ltr">__doc__</code> دیده می‌شود.<br><br><h3 style="margin-top:0;">خطاهای رایج در توابع</h3><ul><li><b>فراموش کردن <code class="ltr">return</code>:</b> اگر چیزی برنگردانید، خروجی تابع <code class="ltr">None</code> است.</li><li><b>استفاده از متغیر بیرون بدون فهم scope:</b> متغیرهای داخل تابع محلی‌اند؛ اگر بیرون را لازم دارید، آن را به تابع بدهید.</li><li><b>پارامتر پیش‌فرضِ mutable:</b> مثل <code class="ltr">def f(x=[])</code> می‌تواند باگ بسازد؛ بهتر است <code class="ltr">None</code> بگذارید و داخل تابع لیست بسازید.</li><li><b>اسم‌گذاری بد:</b> اسم‌های مبهم مثل <code class="ltr">a</code>/<code class="ltr">b</code> در پروژه‌ها سریع گیج‌کننده می‌شوند.</li></ul></div>

```python
def to_meter(cm):
    """سانتی‌متر را به متر تبدیل می‌کند."""
    return cm / 100
```

<div dir="rtl" style="text-align:right;"><h2 style="margin-top:0;">پروژه کوچک (جمع‌بندی): دفترچه نمرات ساده</h2>ایده: کاربر نام دانشجو و نمره را وارد می‌کند، ما در یک دیکشنری ذخیره می‌کنیم و در پایان میانگین می‌گیریم.<br>در این پروژه از حلقه، دیکشنری، تابع و <code class="ltr">try/except</code> استفاده می‌کنیم.<br></div>

```python
def add_score(db, name, score):
    db[name] = score

def average_score(db):
    if not db:
        return 0
    return sum(db.values()) / len(db)

scores = {}

while True:
    name = input("نام (exit برای پایان): ")
    if name.lower() == "exit":
        break

    try:
        score = float(input("نمره: "))
        add_score(scores, name, score)
    except ValueError:
        print("نمره باید عدد باشد!")

print("داده‌ها:", scores)
print("میانگین:", average_score(scores))
```
