```python
try:
    num1 = int(input('Enter a number: '))
    num2 = int(input('Enter another number: '))
    num3 = num1 / num2
    print(num3)
except ZeroDivisionError:
    print('zero error')
except ValueError:
    print('error')
else:
    print('برنامه بدون هیچ مشکلی اجرا شد')
finally:
    print("برنامه به اتمام رسید")
```

```python
while True:
    try:
        num1 = input('Enter a number: / for exit type exit')
        if num1 == 'exit':
            break
        num1 = int(num1)
        num2 = int(input('Enter another number: '))

        num3 = num1 / num2
        print(num3)
        break
    except ZeroDivisionError:
        print('zero error')
    except ValueError:
        print('error')
    finally:
        print("برنامه به اتمام رسید")
```
